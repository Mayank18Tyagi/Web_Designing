## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

````python
# Quantum vs Classical Dice Simulation
# Requires: qiskit, qiskit-aer

import random
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# ----------------------------
# Classical Dice
# ----------------------------
def classical_dice():
    return random.randint(1,6)


# ----------------------------
# Quantum Dice using 3 Qubits
# ----------------------------
def quantum_dice():

    # Create quantum circuit with 3 qubits
    qc = QuantumCircuit(3,3)

    # Put qubits in superposition
    qc.h([0,1,2])

    # Measure qubits
    qc.measure([0,1,2],[0,1,2])

    # Run simulation
    simulator = AerSimulator()
    compiled = transpile(qc, simulator)
    result = simulator.run(compiled, shots=1).result()

    counts = result.get_counts()
    outcome = list(counts.keys())[0]

    # Convert binary to decimal
    value = int(outcome,2)

    # Map result to 1–6
    dice_value = (value % 6) + 1

    return dice_value


# ----------------------------
# Compare Dice
# ----------------------------
print("Rolling Dice...\n")

for i in range(10):
    c = classical_dice()
    q = quantum_dice()

    print(f"Roll {i+1}: Classical Dice = {c} | Quantum Dice = {q}")
````

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

def quantum_dice_roll():

    # STEP 1: Create circuit (3 qubits, 3 classical bits)
    qc = QuantumCircuit(3,3)

    # STEP 2: Apply Hadamard gates to create superposition
    qc.h(0)
    qc.h(1)
    qc.h(2)

    # STEP 3: Measure the qubits
    qc.measure([0,1,2], [0,1,2])

    # STEP 4: Run the circuit once (1 dice roll)
    simulator = AerSimulator()
    job = simulator.run(qc, shots=1)

    result = job.result()
    counts = result.get_counts()

    # Example outcome: '101'
    outcome = list(counts.keys())[0]

    # STEP 5: Convert binary → decimal
    decimal_value = int(outcome, 2)

    # STEP 6: Use numbers 1–6 only
    if decimal_value < 6:
        dice_number = decimal_value + 1
    else:
        # If result is 6 or 7 → reroll
        return quantum_dice_roll()

    print(f"Quantum outcome (binary): {outcome}")
    print(f"Decimal value: {decimal_value}")
    print(f"🎲 Final dice roll = {dice_number}\n")

    return dice_number


# --- Play the game ---
quantum_dice_roll()
````

````python
# ----- CONSTANT BAGS -----

def constant_oracle_always_zero(x):
    # CONSTANT bag: all balls are GREEN. Ignores x completely.
    # Like a broken stamp machine that always says APPROVED.
    return 0      # colour never depends on x at all


def constant_oracle_always_one(x):
    # CONSTANT bag: all balls are RED. Ignores x completely.
    # Like a broken stamp machine that always says REJECTED.
    return 1      # colour never depends on x at all


# ----- BALANCED BAGS -----
# These bags use the ball NUMBER to decide the colour.
# Exactly half the balls are green, exactly half are red.


def balanced_oracle_parity(x):
    # BALANCED bag: uses PARITY of the ball number.
    # Parity = count of '1' bits in binary representation, mod 2.
    # even number of 1-bits -> 0 (GREEN)
    # odd number of 1-bits  -> 1 (RED)

    return bin(x).count('1') % 2


def balanced_oracle_msb(x, n_bits=3):
    # BALANCED bag using most significant bit
    return (x >> (n_bits - 1)) & 1

# ---- VERIFY ALL FOUR ORACLES ----

print("Verifying oracle balance for n=3 bits (8 balls):")
print()

n_test = 3
N = 2 ** n_test

oracles = [
    ("constant_always_zero", constant_oracle_always_zero, ()),
    ("constant_always_one", constant_oracle_always_one, ()),
    ("balanced_parity", balanced_oracle_parity, ()),
    ("balanced_msb", balanced_oracle_msb, (n_test,))
]

for name, fn, args in oracles:
    outputs = [fn(x, *args) for x in range(N)]

    greens = outputs.count(0)
    reds = outputs.count(1)

    verdict = "CONSTANT" if len(set(outputs)) == 1 else "BALANCED"

    print(f"{name:<26}: {greens} green {reds} red -> {verdict}")

print()
print("All four oracles verified correct!")
````

````python
import time


def classical_deutsch_jozsa(oracle, n_bits, verbose=False):
    """
    Decide CONSTANT or BALANCED by pulling balls one at a time.

    Parameters
    ----------
    oracle  : callable  the black-box function f(x) -> {0, 1}
    n_bits  : int       input bit count (bag has 2^n balls)
    verbose : bool      print each pull (great for small demos)
    """

    n_inputs = 2 ** n_bits              # total balls in bag (e.g. n=3 -> 8 balls)
    max_pulls = n_inputs // 2 + 1       # if still one colour after this many pulls -> CONSTANT
                                        # (checking half+1 is enough to be 100% certain)

    seen_colours = set()                # stores which colours (0=green,1=red) we've seen so far
    q = 0                               # counts how many balls we've pulled (queries made)

    t0 = time.perf_counter()            # start timer to measure how long this takes

    for ball_number in range(max_pulls):    # pull balls one by one up to the worst-case limit

        colour = oracle(ball_number)        # ask oracle: what colour is this ball? (0=green,1=red)

        q += 1                              # increment query counter
        seen_colours.add(colour)            # record this colour as seen

        if verbose:
            print(f"pull {ball_number:2d}  -> colour = {colour}")

        # RULE 1 — EARLY EXIT
        # If we've seen BOTH (0 green) AND (1 red), the function is BALANCED.
        # No need to pull more balls — stop immediately and save time.

        if 0 in seen_colours and 1 in seen_colours:
            ms = (time.perf_counter() - t0) * 1000
            return "BALANCED", q, ms

    # RULE 2 — FULL SCAN COMPLETE
    # Pulled N/2+1 balls and only ever saw ONE colour -> must be CONSTANT.

    ms = (time.perf_counter() - t0) * 1000
    return "CONSTANT", q, ms


print("Classical algorithm defined!")

# ============================================================
# STEP-BY-STEP DEMO (n = 2, verbose mode)
# ============================================================

print("=" * 55)
print("STEP-BY-STEP DEMO: n=2 bits (just 4 balls: 0, 1, 2, 3)")
print("=" * 55)

print("\nTest 1: BALANCED bag (parity oracle)")
result, queries, time_taken = classical_deutsch_jozsa(balanced_oracle_parity, 3, verbose=True)

print("\nResult:", result)
print("Oracle queries used:", queries)
print(f"Execution time: {time_taken:.4f} ms")
print("Classical Time Complexity ≈ O(2^n)")
print("Worst-case queries =", (2**3)//2 + 1)


print("\nTest 2: CONSTANT bag (always-zero oracle)")
result, queries, time_taken = classical_deutsch_jozsa(constant_oracle_always_zero, 4, verbose=True)

print("\nResult:", result)
print("Oracle queries used:", queries)
print(f"Execution time: {time_taken:.4f} ms")
print("Classical Time Complexity ≈ O(2^n)")
print("Worst-case queries =", (2**4)//2 + 1)

print("\n(No early exit — needed all N/2 + 1 pulls)")
````

````python
# ===============================
# IMPORTS
# ===============================
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator


# ===============================
# CONSTANT ORACLE CIRCUIT
# ===============================
def build_dj_constant_circuit(n, output_value=0):

    total = n + 1
    qc = QuantumCircuit(total, n)

    # Prepare ancilla in |1>
    qc.x(n)

    # Create superposition on all qubits
    qc.h(range(total))

    # Oracle
    if output_value == 1:
        qc.x(n)

    # Interference step
    qc.h(range(n))

    # Measure inputs
    qc.measure(range(n), range(n))

    return qc


# ===============================
# BALANCED ORACLE CIRCUIT
# ===============================
def build_dj_balanced_circuit(n):

    total = n + 1
    qc = QuantumCircuit(total, n)

    # Prepare ancilla in |1>
    qc.x(n)

    # Create superposition
    qc.h(range(total))

    # Balanced oracle using parity
    for i in range(n):
        qc.cx(i, n)

    # Interference step
    qc.h(range(n))

    # Measure inputs
    qc.measure(range(n), range(n))

    return qc


# ===============================
# RUN CIRCUIT HELPER
# ===============================
def run_circuit(qc, shots=1024):

    simulator = AerSimulator()

    compiled = transpile(qc, simulator)

    job = simulator.run(compiled, shots=shots)

    result = job.result()

    return result.get_counts()


# ===============================
# BUILD CIRCUITS
# ===============================
n = 3

qc_const0 = build_dj_constant_circuit(n, 0)
qc_const1 = build_dj_constant_circuit(n, 1)
qc_balanced = build_dj_balanced_circuit(n)


# ===============================
# RUN ALL CIRCUITS
# ===============================
shots = 1024

print(f"Running quantum circuits on AerSimulator (n={n}, shots={shots})")
print()

counts_const0 = run_circuit(qc_const0, shots=shots)
counts_const1 = run_circuit(qc_const1, shots=shots)
counts_balanced = run_circuit(qc_balanced, shots=shots)

print("Results (1024 shots each):")
print(f" Constant-0 oracle (all-green bag): {counts_const0}")
print(f" Constant-1 oracle (all-red bag): {counts_const1}")
print(f" Balanced oracle (half-half bag): {counts_balanced}")
print()
````

````python
from qiskit import QuantumCircuit, transpile
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler

# connect to IBM Quantum
service = QiskitRuntimeService()

backend = service.backend("ibm_torino")
print("Using backend:", backend.name)

# -------------------------
# CONSTANT ORACLE
# -------------------------
def constant_oracle(n):

    qc = QuantumCircuit(n+1, n)

    qc.x(n)
    qc.h(range(n+1))

    qc.h(range(n))
    qc.measure(range(n), range(n))

    return qc


# -------------------------
# BALANCED ORACLE
# -------------------------
def balanced_oracle(n):

    qc = QuantumCircuit(n+1, n)

    qc.x(n)
    qc.h(range(n+1))

    for i in range(n):
        qc.cx(i, n)

    qc.h(range(n))
    qc.measure(range(n), range(n))

    return qc


# build circuits
n = 3
qc_const = constant_oracle(n)
qc_bal = balanced_oracle(n)

# safe transpilation
qc_const = transpile(qc_const, backend, optimization_level=1)
qc_bal = transpile(qc_bal, backend, optimization_level=1)

sampler = Sampler(mode=backend)

job = sampler.run([qc_const, qc_bal], shots=1024)

result = job.result()

print("Constant oracle result:", result[0].data.meas.get_counts())
print("Balanced oracle result:", result[1].data.meas.get_counts())
````

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit_aer.primitives import Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

# 3-qubit GHZ
qc3 = QuantumCircuit(3,3)
qc3.h(0)
qc3.cx(0,1)
qc3.cx(0,2)
qc3.measure([0,1,2],[0,1,2])

# 5-qubit GHZ
qc5 = QuantumCircuit(5,5)
qc5.h(0)

for i in range(1,5):
    qc5.cx(0,i)

qc5.measure(range(5),range(5))


# Simulator
sim_backend = AerSimulator()

pm_ideal = generate_preset_pass_manager(
    backend=sim_backend,
    optimization_level=1
)

isa_qc3_ideal = pm_ideal.run(qc3)
isa_qc5_ideal = pm_ideal.run(qc5)


# Run circuits
sampler_ideal = Sampler()

job3 = sampler_ideal.run([isa_qc3_ideal],shots=1024)
job5 = sampler_ideal.run([isa_qc5_ideal],shots=1024)

counts3_ideal = job3.result().quasi_dists[0]
counts5_ideal = job5.result().quasi_dists[0]


print("3-qubit GHZ (ideal):", counts3_ideal)
print("Should see 000 and 111\n")

print("5-qubit GHZ (ideal):", counts5_ideal)
print("Should see 00000 and 11111")
````

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime.fake_provider import FakeBrisbane
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import Sampler

# ====================================
# STEP 1: Build GHZ Circuits
# ====================================

# ----- 3-qubit GHZ -----
qc3 = QuantumCircuit(3,3)

qc3.h(0)
qc3.cx(0,1)
qc3.cx(0,2)

qc3.measure([0,1,2],[0,1,2])


# ----- 5-qubit GHZ -----
qc5 = QuantumCircuit(5,5)

qc5.h(0)
qc5.cx(0,1)
qc5.cx(0,2)
qc5.cx(0,3)
qc5.cx(0,4)

qc5.measure(range(5), range(5))


# ====================================
# STEP 2: Create Fake Brisbane backend
# ====================================

brisbane_backend = FakeBrisbane()

# create noisy simulator from backend
noisy_sim = AerSimulator.from_backend(brisbane_backend)


# ====================================
# STEP 3: Transpile (hardware mapping)
# ====================================

pm_noisy = generate_preset_pass_manager(
    backend=noisy_sim,
    optimization_level=1
)

isa_qc3_noisy = pm_noisy.run(qc3)
isa_qc5_noisy = pm_noisy.run(qc5)


# ====================================
# STEP 4: Run circuits
# ====================================

sampler_noisy = Sampler(mode=noisy_sim)

counts3_noisy = sampler_noisy.run(
    [isa_qc3_noisy], shots=1024
).result()[0].data.c.get_counts()

counts5_noisy = sampler_noisy.run(
    [isa_qc5_noisy], shots=1024
).result()[0].data.c.get_counts()


# ====================================
# STEP 5: Print Results
# ====================================

print("🔴 3-qubit GHZ (Noisy FakeBrisbane):", counts3_noisy)
print("   → Extra states beyond '000'/'111' = noise errors!")
print()

print("🔴 5-qubit GHZ (Noisy FakeBrisbane):", counts5_noisy)
print("   → More qubits = more errors — Criteria 1 challenge!")
````

````python
# =========================================
# INSTALL VISUALIZATION LIBRARIES (Run once)
# =========================================
# Uncomment if running in Colab/Jupyter
# !pip install qiskit qiskit-aer pylatexenc

# =========================================
# IMPORTS
# =========================================
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# =========================================
# CIRCUIT A : NO RESET
# =========================================
qc_no_reset = QuantumCircuit(1,1)

qc_no_reset.h(0)       # simulate unknown state
qc_no_reset.x(0)       # flip state
qc_no_reset.measure(0,0)

print("Circuit A (No Reset)")
print(qc_no_reset.draw())   # text visualization


# =========================================
# CIRCUIT B : WITH RESET
# =========================================
qc_with_reset = QuantumCircuit(1,1)

qc_with_reset.h(0)     # simulate unknown state
qc_with_reset.reset(0) # reset qubit to |0>
qc_with_reset.x(0)     # flip |0> -> |1>
qc_with_reset.measure(0,0)

print("\nCircuit B (With Reset)")
print(qc_with_reset.draw())


# =========================================
# RUN SIMULATION
# =========================================
simulator = AerSimulator()

job1 = simulator.run(qc_no_reset, shots=1000)
job2 = simulator.run(qc_with_reset, shots=1000)

counts1 = job1.result().get_counts()
counts2 = job2.result().get_counts()

print("\nResults Circuit A (No Reset):", counts1)
print("Results Circuit B (With Reset):", counts2)


# =========================================
# HISTOGRAM VISUALIZATION
# =========================================
plot_histogram([counts1, counts2],
               legend=["No Reset", "With Reset"])

plt.show()
````

````python
# =========================================
# IMPORTS
# =========================================
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# =========================================
# DELAY TIMES (nanoseconds)
# =========================================
delays_ns = [0, 40000, 80000, 120000, 200000, 250000]

sim = AerSimulator()

results = {}

# =========================================
# BUILD AND RUN CIRCUITS
# =========================================
for delay in delays_ns:

    qc = QuantumCircuit(1,1)

    # Prepare excited state |1>
    qc.x(0)   # "ice cube taken out of freezer"

    # Let the qubit relax
    if delay > 0:
        qc.delay(delay, 0, unit='ns')

    # Measure
    qc.measure(0,0)

    job = sim.run(qc, shots=1024)
    counts = job.result().get_counts()

    results[delay] = counts

# =========================================
# PRINT RESULTS
# =========================================
for delay, counts in results.items():
    print(f"Delay {delay} ns -> {counts}")

# =========================================
# PLOT PROBABILITY OF |1>
# =========================================
p1_values = []

for delay in delays_ns:
    counts = results[delay]
    p1 = counts.get('1',0)/1024
    p1_values.append(p1)

plt.plot(delays_ns, p1_values, marker='o')
plt.xlabel("Delay (ns)")
plt.ylabel("Probability of |1>")
plt.title("Ice Cube Experiment (T1 Relaxation)")
plt.grid()
plt.show()
````

````python
# =========================================
# IMPORTS
# =========================================
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# =========================================
# BUILD A 50/50 SUPERPOSITION CIRCUIT
# =========================================

qc_poll = QuantumCircuit(1,1)

qc_poll.h(0)              # Hadamard gate → 50% |0>, 50% |1>
qc_poll.measure(0,0)

print("Quantum Circuit:")
print(qc_poll.draw())

# =========================================
# SIMULATOR
# =========================================

sim = AerSimulator()

# Different "poll sizes"
shot_sizes = [5, 10, 50, 100, 500, 1000, 10000]

results = {}
p1_values = []

# =========================================
# RUN EXPERIMENT
# =========================================

for shots in shot_sizes:

    job = sim.run(qc_poll, shots=shots)
    counts = job.result().get_counts()

    results[shots] = counts

    p1 = counts.get('1',0) / shots
    p1_values.append(p1)

    print(f"{shots:>6} shots -> P(|1>) = {p1*100:.1f}%")

# =========================================
# HISTOGRAM VISUALIZATION
# =========================================

plot_histogram(results[max(shot_sizes)])
plt.title("Qubit Measurement Distribution")
plt.show()

# =========================================
# EXIT POLL ACCURACY GRAPH
# =========================================

plt.plot(shot_sizes, p1_values, marker='o')
plt.axhline(0.5, linestyle='--')
plt.xlabel("Number of Shots (Poll Size)")
plt.ylabel("Measured Probability of |1>")
plt.title("Exit Poll on Qubit: Sampling Accuracy")
plt.xscale('log')
plt.grid()
plt.show()
````
